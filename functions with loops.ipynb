```python
#-----------------------------------------
#Problem 3: Student Result Processing System
#-----------------------------------------

def process_result(mark_list):
    total_marks = 0
    for mark in mark_list:
        total_marks += mark
    if len(mark_list) == 0:
        return "No Marks Available"

    avg_marks = total_marks / len(mark_list)
    if avg_marks >= 50:
        return "Pass"
    else:
        return "Fail"
    marks_data = [60, 45, 70, 55]
print("Result:", process_result(marks_data))

```

```python
#-----------------------------------------
#Problem 4: Cab Fare Estimator
#-----------------------------------------
def calculate_fare(distance_km, is_peak_time):
    base_fare = 50
    running_charge = 12 * distance_km
    total_amount = base_fare + running_charge

    if is_peak_time:
        total_amount = total_amount * 1.25

    return total_amount


# Sample test cases
print("Cab Fare Examples:")

fare1 = calculate_fare(10, True)
print("Distance: 10 km | Peak: Yes → ₹", round(fare1, 2))

fare2 = calculate_fare(5, False)
print("Distance: 5 km | Peak: No → ₹", round(fare2, 2))

fare3 = calculate_fare(15, True)
print("Distance: 15 km | Peak: Yes → ₹", round(fare3, 2))
```

```python
def check_attendance(attendance_list):
    present = 0

    for day in attendance_list:
        if day == "P":
            present += 1

    attendance_percent = (present / len(attendance_list)) * 100

    if attendance_percent >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
    attendance_data = ["P", "P", "A", "P", "P"]
print("Attendance Status:", check_attendance(attendance_data))

```

```python
#-----------------------------------------
#Problem 6: Password Strength Checker
#----------------------------------------

def check_password(password_text):
    allowed_special_chars = "@#$" # Keeping it limited intentionally

    digit_found = False
    special_found = False

    if len(password_text) < 8:
        return "Weak Password"

    for character in password_text:
        if character.isdigit():
            digit_found = True
        if character in allowed_special_chars:
            special_found = True
            if digit_found and special_found:
                return "Strong Password"
            else:
                return "Weak Password"
print("Password Check:", check_password("Pass@123"))
```
